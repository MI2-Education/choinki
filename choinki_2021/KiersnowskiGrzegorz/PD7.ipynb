```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
X = (np.random.rand(10000)*2-1)
Y = (np.random.rand(10000)*2-1)
Z = np.random.rand(10000)*4
from_middle = np.sqrt(X**2+Y**2)
tab = np.array((X,Y,Z,from_middle))
def color(tab):
    new_array= np.zeros(np.shape(tab)[1], dtype=object)
    new_array[(tab[2]<1) & (tab[3]<0.2)]='brown'
    new_array[(tab[2]>=1)& (tab[2]<2) & (tab[3]<2-tab[2])]='green'
    new_array[(tab[2]>=2)& (tab[2]<2.7) & (tab[3]<2.7-tab[2])]='green'
    new_array[(tab[2]>=2.7)& (tab[2]<3.2) & (tab[3]<3.2-tab[2])]='green'
    new_array[(tab[2]>=3.2)& (tab[2]<3.5) & (tab[3]<3.5-tab[2])]='green'
    new_array[(tab[2]>=1)& (tab[2]<2) & (tab[3]>=2-tab[2]) & (tab[3]<2.03-tab[2])]='blue'
    new_array[(tab[2]>=2)& (tab[2]<2.7) & (tab[3]>=2.7-tab[2])& (tab[3]<2.73-tab[2])]='red'
    new_array[(tab[2]>=2.7)& (tab[2]<3.2) & (tab[3]>=3.2-tab[2])& (tab[3]<3.23-tab[2])]='blue'
    new_array[(tab[2]>=3.2)& (tab[2]<3.5) & (tab[3]>=3.5-tab[2])& (tab[3]<3.53-tab[2])]='red'
    new_array[(tab[2]>=1)& (tab[2]<2) & (tab[3]>=2-tab[2]) & (tab[3]<2.015-tab[2])]='lime'
    new_array[(tab[2]>=2)& (tab[2]<2.7) & (tab[3]>=2.7-tab[2])& (tab[3]<2.715-tab[2])]='yellow'
    new_array[(tab[2]>=2.7)& (tab[2]<3.2) & (tab[3]>=3.2-tab[2])& (tab[3]<3.215-tab[2])]='lime'
    new_array[(tab[2]>=3.2)& (tab[2]<3.5) & (tab[3]>=3.5-tab[2])& (tab[3]<3.515-tab[2])]='yellow'
    return new_array
color_array=color(tab)
```

```python
%matplotlib inline
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_zlim(0,5)
plt.xlim(-3,3)
plt.ylim(-3,3)
ax.view_init(10, 63)

for i in range(10000):# choinka
    if color_array[i]!=0:
        ax.scatter(X[i], Y[i], Z[i], color=color_array[i])
for i in range(20): # gwiazdka
    rd = np.random.rand(3)/5-0.1
    ax.scatter(rd[0], rd[1], rd[2]+3.6, color= 'goldenrod')
for x in np.round(np.arange(1,2.1,0.1), 1):# prezent
    for y in np.round(np.arange(1,2.1,0.1), 1):
        for z in np.arange(0.0,1.1,0.1):
            if x in(1,2) or y in (1,2) or z in (0,1): # prezenty sÄ… puste w Å›rodku tak jak typowy student InÅ¼ynierii i Analizy Danych programujÄ…cy w Å›wiÄ™ta
                if 1.4<=x<=1.6 or 1.4<=y<=1.6 or 0.4<=z<=0.6:
                    ax.scatter(x/2+0.5,y/2+0.5,z/2, color='yellow')
                else:
                    ax.scatter(x/2+0.5,y/2+0.5,z/2, color='red')
for x in np.round(np.arange(1,2.1,0.1), 1):# drugi prezent
    for y in np.round(np.arange(-1,-2.1,-0.1), 1):
        for z in np.arange(0.0,1.1,0.1):
            if x in(1,2) or y in (-1,-2) or z in (0,1):
                if 1.4<=x<=1.6 or -1.4>=y>=-1.6 or 0.4<=z<=0.6:
                    ax.scatter(x/2+0.5,y/2-0.5,z/2, color='lime')
                else:
                    ax.scatter(x/2+0.5,y/2-0.5,z/2, color='blue')
```
