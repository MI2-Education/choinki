```python
# dane : https://gist.github.com/DeastinY/899d532069febdb969d50eb68b7be583

with open('input.txt', encoding = 'utf-8') as file:
    lines = [line.rstrip() for line in file]
```

```python
filtered = filter(lambda line: len(line) > 2, lines)

lines = list(filtered)
```

```python
flattened = []

text = ""

for _ in lines:
    slowa = _.split(" ")
    text += " ".join(slowa)
    text += " "
```

```python
znaki = ",./<>?;':\"[]}{\\|=+-_)(*&^%$#@!`~"

for _ in znaki:
    text = text.replace(_, "")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100).generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image

christmas_tree = np.array(Image.open("christmastree.jpg"))
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=christmas_tree, contour_width=1.5, contour_color='green')

wc.generate(text)

# show
plt.figure(figsize=[50,25])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud.png')
plt.show()
```
