```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.rcParams['figure.figsize'] = (35, 13)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
```

```python
def init_state():
    Z = [i for i in range(360)]
    X = [math.cos(i/4)*(360-i) for i in range(360)]
    Y = [math.sin(i/4)*(360-i) for i in range(360)]
    ax.scatter(X,Y,Z, c="darkgreen",s=20)

    Z = [i for i in range(0,360,9)]
    X = [math.cos(i/4 + 3)*(360-i) for i in range(0,360,9)]
    Y = [math.sin(i/4 + 3)*(360-i) for i in range(0,360,9)]
    ax.scatter(X,Y,Z,c = "orange", marker = "*",s=200)
    c = [(i/360,abs(0.5-i/360),i/360) for i in range(1,360,9)]
    ax.scatter(X,Y,Z, c=c, marker="*",s=50)

    Z = [i for i in range(0,360,11)]
    X = [math.cos(i/4 + 2)*(360-i) for i in range(0,360,11)]
    Y = [math.sin(i/4 + 2)*(360-i) for i in range(0,360,11)]
    ax.scatter(X,Y,Z,c = "red", marker = "o",s=100)
    c = [(i/360,abs(0.5-i/360),i/360) for i in range(1,360,11)]
    ax.scatter(X,Y,Z, c=c, marker="o",s=50)
    return fig
```

```python
def animate(f):
    n = 360
    fig.clear()
    ax = fig.add_subplot(111, projection="3d")
    Z = [i for i in range(n)]
    X = [math.cos(i/4+f/5)*(n-i) for i in range(n)]
    Y = [math.sin(i/4+f/5)*(n-i) for i in range(n)]
    ax.scatter(X,Y,Z, c="darkgreen",s=20)
    
    Z = [i for i in range(0,n,9)]
    X = [math.cos(i/5+2+f/5)*(n-i) for i in range(0,n,9)]
    Y = [math.sin(i/5+2+f/5)*(n-i) for i in range(0,n,9)]
    ax.scatter(X,Y,Z,c = "orange", marker = "*",s=200)
    c = [(i/n,abs(0.5-i/n),i/n) for i in range(0,n,9)]
    ax.scatter(X,Y,Z, c=c, marker="*",s=50)
    
    Z = [i for i in range(0,n,11)]
    X = [math.cos(i/4 + 2 + f/5)*(n-i) for i in range(0,n,11)]
    Y = [math.sin(i/4 + 2 + f/5)*(n-i) for i in range(0,n,11)]
    ax.scatter(X,Y,Z,c = "red", marker = "o",s=100)
    c = [(i/n,abs(0.5-i/n),i/n) for i in range(0,n,11)]
    ax.scatter(X,Y,Z, c=c, marker="*",s=50)
    return fig
```

```python
from matplotlib import animation
animation = animation.FuncAnimation(fig,animate,init_func = init_state)
animation.save("drzewko.gif")
```
