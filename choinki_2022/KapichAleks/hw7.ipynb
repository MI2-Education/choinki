```python
from mplsoccer.pitch import VerticalPitch
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./christmas.csv')
df["Start X"] = df["Start X"]*80
df["End X"] = df["End X"]*80
df["Start Y"] = df["Start Y"]*120
df["End Y"] = df["End Y"]*120
df
```

```python
fig ,ax = plt.subplots(figsize=(8,10))
fig.set_facecolor('#192148')
ax.patch.set_facecolor('#192148')

pitch = VerticalPitch(pitch_type='statsbomb',
                     pitch_color='#192148',
                     line_color="white")
pitch.draw(ax=ax)

plt.title('Merry Christmas!', color='white', size=20)

for x in range(len(df['Start X'])):
        plt.plot((df['Start X'][x], df['End X'][x]),
                 (df['Start Y'][x], df['End Y'][x]),
                 color=df['Colour'][x])
plt.scatter(40, 104, color='yellow', marker="*", s=1000)
plt.savefig('choinka.png')
```
